```python
# For EDA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/gdp-growth-of-african-countries/Africa_GDP.csv')
df.head()
```

```python
df.info()
```

```python
df.describe().T
```

```python
def outlier_check(row, colname):  
    Q1 = row[colname].quantile(0.25)
    Q3 = row[colname].quantile(0.75)
    IQR = Q3 - Q1
    lower_b = Q1 - 1.5 * IQR
    upper_b = Q3 + 1.5 * IQR
    outlier = row.loc[(row[colname] < lower_b) | (row[colname] > upper_b)]
    return outlier
    
outliers = pd.DataFrame() 

for col in df.select_dtypes(include=['number']).columns: 
    outliers_in_col = outlier_check(df, col) 
    outliers = pd.concat([outliers, outliers_in_col])

outliers.reset_index(drop=True, inplace=True)
outliers
```

```python
print("Missing Value (%)")
missing_values = df.isna().mean() * 100
print(missing_values)

print("\nDuplicate Rows (%):")
duplicate_count = df.duplicated().mean()
print(duplicate_count)
```

# <div style="display: inline-block; padding: 20px 20px; border-radius: 15px; background: linear-gradient(135deg, #d8bfd8, #8b008b); box-shadow: 3px 3px 8px rgba(0, 0, 0, 0.2); font-size: 35px; font-family: Arial, sans-serif; color: #ffffff; text-align: center;">EDA : Data Distribution</div>

```python
plt.figure(figsize=(15, 8))
for country in df.columns[1:]:
    plt.plot(df["Year"], df[country], label=country)

plt.title("Country Data Over Years")
plt.xlabel("Year")
plt.ylabel("Values")
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.boxplot(data=df.iloc[:, 1:])
plt.xticks(rotation=90)
plt.title("Data Distribution by Country")
plt.show()
```

```python
# Specific year

year = 2020  

specific_year_data = df[df["Year"] == year].iloc[:, 1:].T
specific_year_data.columns = ["Value"]

specific_year_data.sort_values(by="Value", ascending=False).plot(kind="bar", figsize=(15, 8))
plt.title(f"Data for {year}")
plt.ylabel("Values")
plt.xlabel("Countries")
plt.show()
```

# <div style="display: inline-block; padding: 20px 20px; border-radius: 15px; background: linear-gradient(135deg, #d8bfd8, #8b008b); box-shadow: 3px 3px 8px rgba(0, 0, 0, 0.2); font-size: 35px; font-family: Arial, sans-serif; color: #ffffff; text-align: center;">EDA : Correlation Analysis</div>


```python
plt.figure(figsize=(15, 10))
sns.heatmap(df.iloc[:, 1:].corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Between Countries")
plt.show()
```
