# **Data Visualization**

> ## Trend of Transaction Count by Month, Hour, Day

```python
transaction_counts_month  = data.groupby('Month Name')['transaction_id'].count().reset_index(name='Number of transaction_id')
transaction_counts_hour  = data.groupby('Hour')['transaction_id'].count().reset_index(name='Number of transaction_id')
transaction_counts_week  = data.groupby('Day Name')['transaction_id'].count().reset_index(name='Number of transaction_id')
```

```python
# Trend of Transaction Count by Month
plt.figure(figsize=(10, 6)) 
sns.barplot(x='Month Name', y='Number of transaction_id', data = transaction_counts_month, palette = 'BrBG') 
plt.title('Trend of Transaction Count by Month', fontsize=14)
plt.xlabel('Month', fontsize=14) 
plt.ylabel('Number of transaction_id', fontsize=14) 
plt.xticks(rotation=0, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--')

# Trend of Transaction Count by Hour
plt.figure(figsize=(10, 6)) 
plt.plot('Hour', 'Number of transaction_id', data=transaction_counts_hour, marker='o', label='Transaction Count') 
plt.title('Trend of Transaction Count by Hour', fontsize=14)
plt.xlabel('Hour', fontsize=14) 
plt.ylabel('Number of transaction_id', fontsize=14) 
plt.xticks(rotation=0, fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=12)  
plt.grid(axis='y', linestyle='--')
plt.show()

# Trend of Transaction Count by Week
plt.figure(figsize=(10, 6)) 
sns.barplot(x='Day Name', y='Number of transaction_id', data = transaction_counts_week, palette = 'coolwarm') 
plt.title('Trend of Transaction Count by Week', fontsize=14)
plt.xlabel('Day', fontsize=14) 
plt.ylabel('Number of transaction_id', fontsize=14) 
plt.xticks(rotation=0, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--')

plt.show()
```

> ## Trend of Bill by Month, Hour, Day

```python
Bill_month  = data.groupby('Month Name')['Total_Bill'].sum().reset_index(name='Total_Bill')
Bill_hour  = data.groupby('Hour')['Total_Bill'].sum().reset_index(name='Total_Bill')
Bill_week  = data.groupby('Day Name')['Total_Bill'].sum().reset_index(name='Total_Bill')
```

```python
# Trend of Total_Bill (Month, Hour, Day)

plt.figure(figsize=(10, 6)) 
sns.barplot(x='Month Name', y='Total_Bill', data = Bill_month, palette = 'BrBG') 
plt.title('Trend of Total_Bill', fontsize=14)
plt.xlabel('Month', fontsize=14) 
plt.ylabel('Total_Bill', fontsize=14) 
plt.xticks(rotation=0, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--')

plt.figure(figsize=(10, 6)) 
plt.plot('Hour', 'Total_Bill', data=Bill_hour, marker='o', label='Total_Bill') 
plt.title('Trend of Total_Bill', fontsize=14)
plt.xlabel('Hour', fontsize=14) 
plt.ylabel('Total_Bill', fontsize=14) 
plt.xticks(rotation=0, fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=12)  
plt.grid(axis='y', linestyle='--')
plt.show()

plt.figure(figsize=(10, 6)) 
sns.barplot(x='Day Name', y='Total_Bill', data = Bill_week, palette = 'coolwarm') 
plt.title('Trend of Total_Bill', fontsize=14)
plt.xlabel('Day', fontsize=14) 
plt.ylabel('Total_Bill', fontsize=14) 
plt.xticks(rotation=0, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--')

plt.show()
```

> ## Trend of Transaction Count by Store

```python
transaction_counts_location  = data.groupby('store_location')['transaction_id'].count().reset_index(name='Number of transaction_id')
transaction_counts_location_sum = transaction_counts_location['Number of transaction_id']
```

```python
plt.pie(transaction_counts_location_sum, labels=transaction_counts_location['store_location'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Accent.colors)
plt.title('Transaction Count by Store Location', fontsize=14)

plt.show()
```

> ## Trend of Total Bill by Store

```python
Bill_location  = data.groupby('store_location')['Total_Bill'].sum().reset_index(name='Total_Bill')
Bill_location_sum = Bill_location['Total_Bill']
```

```python
plt.pie(Bill_location_sum, labels=transaction_counts_location['store_location'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Accent.colors)
plt.title('Total_Bill by Store Location', fontsize=14)

plt.show()
```

> ## Trend of Transaction Count by product

```python
transaction_counts_product  = data.groupby('product_category')['transaction_id'].count().reset_index(name='Number of transaction_id')
```

```python
plt.figure(figsize=(10, 6)) 
sns.barplot(x='product_category', y='Number of transaction_id', data = transaction_counts_product, palette = 'viridis') 
plt.title('Trend of Transaction Count by product', fontsize=14)
plt.xlabel('product_category', fontsize=14) 
plt.ylabel('Number of transaction_id', fontsize=14) 
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--')

plt.show()
```

> ## Trend of Bill by product

```python
Bill_product  = data.groupby('product_category')['Total_Bill'].sum().reset_index(name='Total_Bill')
```

```python
plt.figure(figsize=(10, 6)) 
sns.barplot(x='product_category', y='Total_Bill', data = Bill_product, palette = 'RdBu') 
plt.title('Trend of Bill by product', fontsize=14)
plt.xlabel('product_category', fontsize=14) 
plt.ylabel('Number of transaction_id', fontsize=14) 
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--')

plt.show()
```

# **Data Analysis**

> ## Top 3 : Transaction Count by Month, Hour, Day

```python
plt.figure(figsize=(12, 6))
top_models = data['Month Name'].value_counts().head(3)  
sns.barplot(x=top_models.index, y=top_models.values, palette='viridis')  
plt.title('Top 3 : Month', fontsize=16) 
plt.xlabel('Month', fontsize=14)  
plt.ylabel('Transaction Count', fontsize=14) 
plt.xticks(rotation=0, fontsize=12)  
plt.yticks(fontsize=12)  
plt.grid(axis='y', linestyle='--')  
plt.tight_layout()  

plt.figure(figsize=(12, 6))
top_models = data['Hour'].value_counts().head(3)
sns.barplot(x=top_models.index, y=top_models.values, palette='Set2')  
plt.title('Top 3 : Hour', fontsize=16) 
plt.xlabel('Hour', fontsize=14)  
plt.ylabel('Transaction Count', fontsize=14) 
plt.xticks(rotation=0, fontsize=12)  
plt.yticks(fontsize=12)  
plt.grid(axis='y', linestyle='--')  
plt.tight_layout()  

plt.figure(figsize=(12, 6))
top_models = data['Day Name'].value_counts().head(3)  
sns.barplot(x=top_models.index, y=top_models.values, palette='Set3')  
plt.title('Top 3 : Day', fontsize=16) 
plt.xlabel('Day', fontsize=14)  
plt.ylabel('Transaction Count', fontsize=14) 
plt.xticks(rotation=0, fontsize=12)  
plt.yticks(fontsize=12)  
plt.grid(axis='y', linestyle='--')  
plt.tight_layout()  

plt.show()
```

> ## Top 3 : Transaction Count by Product

```python
plt.figure(figsize=(12, 6))
top_models = data['product_category'].value_counts().head(3)  
sns.barplot(x=top_models.index, y=top_models.values, palette='viridis')  
plt.title('Top 3 : Product', fontsize=16) 
plt.xlabel('Product_categroy', fontsize=14)  
plt.ylabel('Transaction Count', fontsize=14) 
plt.xticks(rotation=0, fontsize=12)  
plt.yticks(fontsize=12)  
plt.grid(axis='y', linestyle='--')  
plt.tight_layout()  
```

> ## Top 3 : Bill by Month, Hour, Day

```python
monthly_max_sales = data.groupby('Month Name')['Total_Bill'].sum()
top_models = monthly_max_sales.sort_values(ascending=False).head(3) 
plt.figure(figsize=(12, 6)) 
sns.barplot(x=top_models.index, y=top_models.values, palette='viridis')  
plt.title('Top 3 : Month', fontsize=16) 
plt.xlabel('Month', fontsize=14)  
plt.ylabel('Total_Bill', fontsize=14) 
plt.xticks(rotation=0, fontsize=12)  
plt.yticks(fontsize=12)  
plt.grid(axis='y', linestyle='--')  
plt.tight_layout()  

monthly_max_sales = data.groupby('Hour')['Total_Bill'].sum()
top_models = monthly_max_sales.sort_values(ascending=False).head(3) 
plt.figure(figsize=(12, 6)) 
sns.barplot(x=top_models.index, y=top_models.values, palette='Set2')  
plt.title('Top 3 : Hour', fontsize=16) 
plt.xlabel('Hour', fontsize=14)  
plt.ylabel('Total_Bill', fontsize=14) 
plt.xticks(rotation=0, fontsize=12)  
plt.yticks(fontsize=12)  
plt.grid(axis='y', linestyle='--')  
plt.tight_layout()  

monthly_max_sales = data.groupby('Day Name')['Total_Bill'].sum()
top_models = monthly_max_sales.sort_values(ascending=False).head(3) 
plt.figure(figsize=(12, 6)) 
sns.barplot(x=top_models.index, y=top_models.values, palette='Set3')  
plt.title('Top 3 : Day', fontsize=16) 
plt.xlabel('Day Name', fontsize=14)  
plt.ylabel('Total_Bill', fontsize=14) 
plt.xticks(rotation=0, fontsize=12)  
plt.yticks(fontsize=12)  
plt.grid(axis='y', linestyle='--')  
plt.tight_layout()  
```
